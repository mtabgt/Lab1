```python
def histograma(image_path):
    img=cv2.imread(image_path)
    if img is None:
        print("Error, link a imagen incorrecto o nombre mal escrito.")
        return
    img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    
    alto=img.shape[0]
    ancho=img.shape[1]
    resultante_salida_gray=np.zeros((alto, ancho,3)) ### Se consigue espacio vacío en el que se va guardando cada resultado de la nueva imagen. Son tres canales debido a RGB
    for i in range(0,alto):
        for j in range(0,ancho):
            pixel=img[i,j]
            blue=pixel[0]
            green=pixel[1]
            red=pixel[2]
            resultante_salida_gray[i,j]=[(blue+green+red)/3]
    
    cv2.imwrite("resultante_salida_gray.jpg", resultante_salida_gray) ### Es necesario hacer este proceso de convertir a imagen lo que se trató y bajarla, y luego volverla a subir, ya que si no puede dar problema en la lectura
    resultante_salida_gray=cv2.imread("resultante_salida_gray.jpg")

    canal_azul, canal_verde, canal_rojo=cv2.split(img)
    media_azul=np.mean(canal_azul)
    media_verde=np.mean(canal_verde)
    media_rojo=np.mean(canal_rojo)
    media_gris=np.mean(resultante_salida_gray)
    
    hist_azul=cv2.calcHist([img],[0], None, [256], [0,256])
    hist_verde=cv2.calcHist([img],[1], None, [256], [0,256])
    hist_rojo=cv2.calcHist([img],[2], None, [256], [0,256])
    hist_grises=cv2.calcHist([resultante_salida_gray],[0], None, [256], [0,256])
    


    plt.figure()
    plt.subplot(4,1,1)
    plt.plot(hist_azul, color="blue")
    plt.axvline(media_azul, color="blue", linestyle="--", label=f"Media:{media_azul:.2f}")
    plt.xlabel("Intensidad de pixeles en azul")
    plt.ylabel("Pixeles totales")
    plt.title("Histograma de imagen original")

    plt.subplot(4,1,2)
    plt.plot(hist_verde, color="green")
    plt.axvline(media_verde, color="green", linestyle="--", label=f"Media:{media_verde:.2f}")
    plt.xlabel("Intensidad de pixeles en verde")
    plt.ylabel("Pixeles totales")
    plt.title("Histograma de imagen original")

    plt.subplot(4,1,3)
    plt.plot(hist_rojo, color="red")
    plt.axvline(media_rojo, color="red", linestyle="--", label=f"Media:{media_rojo:.2f}")
    plt.xlabel("Intensidad de pixeles en rojo")
    plt.ylabel("Pixeles totales")
    plt.title("Histograma de imagen original")

    plt.subplot(4,1,4)
    plt.plot(hist_grises, color="gray")
    plt.axvline(media_gris, color="gray", linestyle="--", label=f"Media:{media_gris:.2f}")
    plt.xlabel("Intensidad de pixeles en gris")
    plt.ylabel("Pixeles totales")
    plt.title("Histograma de imagen original")

    plt.tight_layout
    plt.show()
```

```python
print("Sistema de generación de histogramas de color")
image_path=input("Por favor ingresar la ruta de la imagen que se desea cargar.")
histograma(image_path)
```
